<h1>Soal 2: Pemahaman Machine Learning</h1>

Pelajarilah secara garis besar suatu model/algoritma machine learning, kemudian aplikasikan untuk membuat model bagi data (variable x, y) di bawah ini. Kemudian buatlah prediksi terhadap data training dan data baru di interval 20-30.

- Plot data asli, data hasil prediksi terhadap data training, data hasil prediksi terhadap data baru
- Tunjukan beberapa parameter yang dimiliki model yang telah kalian buat

```python
import matplotlib.pyplot as plt
import numpy as np

rng = np.random.RandomState(42)
x = 20 * rng.rand(50)
y = x**2 + 2 * x + - 1 + rng.randn(50)

# data yang akan diprediksi
x_new = np.arange(20, 30, 0.5)

fig, ax = plt.subplots(figsize=(12, 6))
ax.scatter(x, y)
ax.set_xlabel('X-Value')
ax.set_ylabel('Y-Value')
plt.show()
```

```python
# 1. pilih model
from sklearn.preprocessing import PolynomialFeatures
```

```python
# 2. pilih hyperparameter
model = PolynomialFeatures(2)
```

```python
# 3. pisahkan data feature dan target 

# data feature: x dan y
# data target: x_mat
 
x_mat = x[:, np.newaxis]
```

```python
# 4. perintahkan model untuk mempelajari dengan method fit
model.fit(x_mat, y)
```

```python
# 5. aplikasikan ke dalam test data: x_new, y_new
x_new = x_new[:, np.newaxis]

y_new = model.transform(x_new)

# training data
y_mat = model.transform(x_mat)
```

```python
# plot

fig, ax = plt.subplots(figsize=(12, 6))
ax.scatter(x, y, color = 'green', label = 'Data Asli')
plt.plot(x_new, y_new[:,2], linewidth = 4,color = 'red', label = 'Data Test')
ax.scatter(x_mat, y_mat[:,2], color = 'red', label = 'Data Training')
ax.set_xlabel('X-Value')
ax.set_ylabel('Y-Value')
plt.legend()
plt.show()
```

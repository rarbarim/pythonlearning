```python
import matplotlib.pyplot as plt
import numpy as np 
```

<h2>Soal 1: Multiple Subplots</h2>

Dengan data di bawah ini buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 200)
tan = np.tan(x)/10
cos = np.cos(x)
sin = np.sin(x)
```

```python
fig, ax = plt.subplots(nrows = 1, ncols = 3, figsize = (16,9), sharey = True)

ax[0].plot(x, tan, label = 'tan x')
ax[0].set_title('Grafik tan x')

ax[1].plot(x, sin, color = 'red', label = 'sin x')
ax[1].set_title('Grafik sin x')

ax[2].plot(x, cos, color = 'green', label = 'cos x')
ax[2].set_title('Grafik cos x')

fig.legend(loc = 'upper center')

plt.show()

```

<h2>Soal 2: Nested Axis</h2>

Dengan data di bawah ini, buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 100)
y = np.cos(x)
y2 = np.cos(x**2)
y3 = np.cos(x**3)
y4 = np.cos(x**4)
y5 = np.cos(x**5)
```

```python
fig, ax = plt.subplots(figsize = (16,9))

ax.plot(x, y)
ax.set_title('Grafik cos x')

plot1 = ax.inset_axes([-0.1,0.9,0.3,0.3])
plot1.plot(x,y)
plot1.plot(x,y5, color = 'gold')
plot1.set_title('Grafik cos x^5')

plot2 = ax.inset_axes([0.9,0.9,0.3,0.3])
plot2.plot(x,y)
plot2.plot(x,y4, color = 'black')
plot2.set_title('Grafik cos x^4')

plot3 = ax.inset_axes([-0.1,-0.1,0.3,0.3])
plot3.plot(x,y)
plot3.plot(x,y2, color = 'red')
plot3.set_title('Grafik cos x^2')

plot4 = ax.inset_axes([0.9,-0.1,0.3,0.3])
plot4.plot(x,y)
plot4.plot(x,y3, color = 'green')
plot4.set_title('Grafik cos x^3')

plt.show()
```
